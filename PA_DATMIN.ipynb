**Informasi Umum**

Bank Customer Churn Dataset adalah kumpulan data yang berisi informasi tentang klien bank dan apakah mereka tetap menjadi pelanggan atau tidak. Dalam konteks bisnis, pemahaman data ini akan membantu bank untuk memahami faktor-faktor yang mempengaruhi tingkat keberhasilan dalam mempertahankan pelanggan mereka dan mengurangi churn rate (tingkat pergantian pelanggan).

Beberapa pertanyaan bisnis yang dapat dijawab dengan menggunakan dataset ini antara lain:

> Apa faktor-faktor yang mempengaruhi keputusan klien untuk tetap menjadi pelanggan bank?

> Bagaimana kinerja cabang bank dalam mempertahankan pelanggan mereka?

> Apakah ada perbedaan dalam tingkat churn rate antara jenis kelamin, usia, dan pendapatan klien?

> Bagaimana bank dapat meningkatkan retention rate (tingkat mempertahankan pelanggan) mereka?

Dengan memahami data ini, bank dapat mengambil tindakan untuk meningkatkan retention rate mereka dan mencegah kehilangan pelanggan, seperti memperbaiki layanan pelanggan, menawarkan produk atau promo yang menarik, atau meningkatkan efisiensi operasional mereka.

Dataset bank churn customer adalah parameter atau indikator keberhasilan suatu proyek analisis data dalam mengatasi permasalahan churn customer di bank. Suatu proyek analisis data dianggap berhasil apabila berhasil mengurangi tingkat churn customer di bank dan meningkatkan loyalitas pelanggan, sehingga dapat meningkatkan pendapatan dan profitabilitas bank. Kriteria keberhasilan ini dapat diukur melalui peningkatan retention rate atau customer retention rate (CRR), peningkatan customer satisfaction score (CSAT), dan peningkatan customer lifetime value (CLTV).

**Informasi Atribut**

> customer_id: Identitas unik dari setiap pelanggan di bank.

> estimated_salary: Estimasi gaji tahunan pelanggan.

> tenure: mengacu pada jumlah bulan atau periode waktu yang dihabiskan oleh pelanggan di bank. 

Dataset ini dapat digunakan untuk memprediksi apakah seorang pelanggan akan meninggalkan bank atau tidak berdasarkan pada atribut-atribut tersebut. Hal ini akan membantu bank untuk mengidentifikasi faktor-faktor yang berkontribusi pada tingkat churn dan memungkinkan mereka untuk mengambil tindakan yang tepat untuk mempertahankan pelanggan mereka.

###C4.5

```python
class Node:
    def __init__(self):
        
        self.right = None
        self.left = None
        
        self.column = None
        self.threshold = None
        
        self.probas = None

        self.depth = None
        
        self.is_terminal = False

class DecisionTreeClassifier:
    def __init__(self, max_depth = 3, min_samples_leaf = 1, min_samples_split = 2): 
        self.max_depth = max_depth
        self.min_samples_leaf = min_samples_leaf
        self.min_samples_split = min_samples_split 
        self.classes = None
        
        self.Tree = None
    
    def nodeProbas(self, y):    
        probas = []
        
        for one_class in self.classes:
            proba = y[y == one_class].shape[0] / y.shape[0]
            probas.append(proba)
        return np.asarray(probas)

    def gini(self, probas):
        return 1 - np.sum(probas**2)
    
    def calcImpurity(self, y):
        return self.gini(self.nodeProbas(y))
    
    def calcBestSplit(self, X, y):
        bestSplitCol = None
        bestThresh = None
        bestInfoGain = -999
        
        impurityBefore = self.calcImpurity(y)
        
        for col in range(X.shape[1]):
            x_col = X[:, col]
            
            for x_i in x_col:
                threshold = x_i
                y_right = y[x_col > threshold]
                y_left = y[x_col <= threshold]
                
                if y_right.shape[0] == 0 or y_left.shape[0] == 0:
                    continue
                    
                impurityRight = self.calcImpurity(y_right)
                impurityLeft = self.calcImpurity(y_left)
                
                infoGain = impurityBefore
                infoGain -= (impurityLeft * y_left.shape[0] / y.shape[0]) + (impurityRight * y_right.shape[0] / y.shape[0])
                
                if infoGain > bestInfoGain:
                    bestSplitCol = col
                    bestThresh = threshold
                    bestInfoGain = infoGain
                    
        if bestInfoGain == -999:
            return None, None, None, None, None, None
        
        
        x_col = X[:, bestSplitCol]
        x_left, x_right = X[x_col <= bestThresh, :], X[x_col > bestThresh, :]
        y_left, y_right = y[x_col <= bestThresh], y[x_col > bestThresh]
        
        return bestSplitCol, bestThresh, x_left, y_left, x_right, y_right
                  
    def buildDT(self, X, y, node):
        if node.depth >= self.max_depth:
            node.is_terminal = True
            return
        
        if X.shape[0] < self.min_samples_split:
            node.is_terminal = True
            return
        
        if np.unique(y).shape[0] == 1:
            node.is_terminal = True
            return
        
        splitCol, thresh, x_left, y_left, x_right, y_right = self.calcBestSplit(X, y)
        
        if splitCol is None:
            node.is_terminal = True
            
        if x_left.shape[0] < self.min_samples_leaf or x_right.shape[0] < self.min_samples_leaf:
            node.is_terminal = True
            return
        
        node.column = splitCol
        node.threshold = thresh
        
        node.left = Node()
        node.left.depth = node.depth + 1
        node.left.probas = self.nodeProbas(y_left)
        
        node.right = Node()
        node.right.depth = node.depth + 1
        node.right.probas = self.nodeProbas(y_right)
        
        self.buildDT(x_right, y_right, node.right)
        self.buildDT(x_left, y_left, node.left)
        
    def fit(self, X, y):
        if type(X) == pd.DataFrame:
            X = np.asarray(X)
        
        self.classes = np.unique(y)
        self.Tree = Node()
        self.Tree.depth = 1
        self.Tree.probas = self.nodeProbas(y)
        
        self.buildDT(X, y, self.Tree)
    
    def predictSample(self, x, node):
        if node.is_terminal:
            return node.probas
        
        if x[node.column] > node.threshold:
            probas = self.predictSample(x, node.right)
        else:
            probas = self.predictSample(x, node.left)
            
        return probas
        
    def predict(self, X):   
        if type(X) == pd.DataFrame:
            X = np.asarray(X)
            
        predictions = []
        for x in X:
            pred = np.argmax(self.predictSample(x, self.Tree))
            predictions.append(pred)
        
        return np.asarray(predictions)

# definisikan model Decission Tree Classifier
dtc = DecisionTreeClassifier(max_depth = 10, min_samples_leaf=2, min_samples_split=2)

# latih model Decission Tree Classifier
dtc.fit(X_train, y_train)

# evalusasi performa model menggunakan data uji
dtc_pred = dtc.predict(X_train)
```

###K-Means

```python
dffifa.head()
```

```python
import numpy as np
import pandas as pd

features = ["Age", "Overall", "Potential","Preferred Foot","Reactions", "BallControl", "ShotPower", "Stamina", "Strength", "Vision", "Composure", "GKHandling", "GKReflexes"]

df_norm = (dffifa[features] - dffifa[features].mean()) / dffifa[features].std()

k = 3 
centroids = df_norm.sample(k, random_state=42)

max_iterations = 100
for i in range(max_iterations):
    distances = pd.DataFrame()
    for j in range(k):
        dist = np.sum((df_norm - centroids.iloc[j])**2, axis=1)
        distances[f'distance_to_{j}'] = dist
    cluster_labels = distances.idxmin(axis=1)
    cluster_labels = cluster_labels.map(lambda x: int(x.lstrip("distance_to_")))

    new_centroids = pd.DataFrame(np.zeros((k, len(features))), columns=features)
    for j in range(k):
        new_centroid = df_norm[cluster_labels == j].mean(axis=0)
        new_centroids.loc[j] = new_centroid
    if centroids.equals(new_centroids):
        break
    centroids = new_centroids

dffifa["cluster"] = cluster_labels

print("Centroids:")
print(centroids)
print("\nCluster sizes:")
print(dffifa["cluster"].value_counts())
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_norm.iloc[:, 0], df_norm.iloc[:, 1], c=cluster_labels, alpha=0.5)
plt.scatter(centroids.iloc[:, 0], centroids.iloc[:, 1], marker='x', s=200, linewidths=3,
            color='r', zorder=10, label='Centroids')
plt.title('Clustering Visualization')
plt.xlabel(features[0])
plt.ylabel(features[1])
plt.legend()
plt.show()
```

**Hasil K-means terdiri dari beberapa informasi, di antaranya:**

> Centroids: menunjukkan titik tengah dari masing-masing klaster, yaitu nilai rata-rata dari setiap fitur dalam klaster tersebut. Semakin berbeda nilai centroid dari klaster yang berbeda, semakin jelas perbedaan antara klaster.

> Cluster sizes: menunjukkan jumlah data yang termasuk dalam masing-masing klaster. Semakin berbeda jumlah data dalam klaster yang berbeda, semakin jelas perbedaan antara klaster.

> Cluster labels: menunjukkan klaster mana yang diassign ke setiap data dalam dataset. Setiap data akan diassign ke klaster dengan centroid terdekat.

Untuk menjabarkan ciri-ciri tiap klaster, dapat dilakukan dengan melihat nilai rata-rata (centroid) dari setiap fitur dalam masing-masing klaster. ##Evaluation

###SSE

```python
SSE = pd.DataFrame(columns=features)
for j in range(k):
    cluster_data = df_norm[cluster_labels == j]
    cluster_SSE = ((cluster_data - centroids.iloc[j])**2).sum()
    SSE.loc[f'Cluster{j}'] = cluster_SSE
print("SSE:", SSE)
```

Semakin rendah nilai SSE, semakin baik model clusteringnya. Berdasarkan output, dapat dilihat bahwa Cluster0 memiliki SSE paling rendah untuk hampir semua fitur, sementara Cluster1 dan Cluster2 memiliki SSE yang lebih tinggi. Oleh karena itu, Cluster0 mungkin merupakan cluster yang paling baik dan homogen dibandingkan dengan Cluster1 dan Cluster2.

###Elbow Method

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertias = []
k_range = range(1, 10)
for k in k_range:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(df_norm)
    inertias.append(model.inertia_)

plt.plot(k_range, inertias)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

 Jumlah klaster yang optimal dipilih pada saat di mana penurunan inertia mulai melambat secara signifikan, sehingga membentuk kurva seperti siku atau "elbow". Dalam kasus ini, kita dapat melihat bahwa "elbow" terletak pada 2 atau 3 klaster, sehingga jumlah klaster optimal adalah 2 atau 3.

###Silhouette Score Method

```python
from sklearn.metrics import silhouette_score

sil_scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_norm)
    sil_scores.append(silhouette_score(df_norm, kmeans.labels_))
    
plt.plot(range(2, 10), sil_scores)
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

Output tersebut menunjukkan grafik Silhouette score dengan jumlah klaster yang bervariasi dari 2 hingga 9. Pada grafik tersebut, semakin tinggi skor Silhouette, semakin baik model clustering. Berdasarkan grafik tersebut, kita dapat melihat bahwa skor Silhouette tertinggi terjadi pada 3 klaster, sehingga jumlah klaster yang paling optimal untuk data ini adalah 3.
